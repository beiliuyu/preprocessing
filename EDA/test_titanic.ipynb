```python
from univariate_eda import UnivariateEDA
from bivariate_eda import BivariateEDA
from groupvariate import GroupvariateEDA
%matplotlib inline 
```

```python
fpath = './data/train.csv'
plotUni = UnivariateEDA(fpath)
plotBi = BivariateEDA(fpath)
plotGroup = GroupvariateEDA(fpath)
```

### 船舱等级与生存

```python
x = 'Survived'
y = 'Pclass'
plotBi.plot_categoty_inside(x,y)
plotBi.plot_counts(y,x,bigSize=0.1)
plotBi.plot_joint(x,y)
plotBi.plot_ordered_bar(x,y)
plotBi.plot_stackHist(y,x)
```
